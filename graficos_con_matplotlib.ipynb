Gráfico de una función cuadrática y una lineal

```python
# Función cuadrática.
def f1(x):
    return 2*(x**2) + 5*x - 2
# Función lineal.
def f2(x):
    return 4*x + 1
# Valores del eje X que toma el gráfico.
x=np.arange(-5,5,0.1)
# Graficar ambas funciones.
plt.plot(x, [f1(i) for i in x],label="Parábola",linewidth=2,color="green")
plt.plot(x, [f2(i) for i in x],label="Recta",linewidth=2,color="orange")
# Establecer el título
plt.title("Gráfica de funciones")
# Establecer el color de los ejes.
plt.axhline(0, color="black")
plt.axvline(0, color="black")
# Establecer las etiquetas de los ejes.
plt.ylabel("Eje_y")
plt.xlabel("Eje_x")
# Limitar los valores de los ejes.
plt.xlim(-5, 5)
plt.ylim(-7.5, 15)
# Guardar gráfico como imágen PNG.
#plt.savefig("output.png")
# Mostrarlo.
plt.legend()
plt.grid()
plt.show()
```

Función polinómica

```python
x=np.arange(-20.,20.,0.1)
```

```python
def f(x):
    return 2*x**3-5*x**2+2*x+4
```

```python
plt.figure(figsize=(10,7))
plt.plot(x, [f(i) for i in x])
plt.axhline(0,color="black")
plt.axvline(0,color="black")
plt.xlim(-5,5)
plt.ylim(-100,100)
plt.xlabel("Eje_X")
plt.ylabel("Eje_Y")
plt.title("Curva polinómica")
plt.grid()
plt.show()
```

Función por partes 
$ f(x)= \mid {x}  \mid
\begin{align}
\begin{cases}
{-x} \text {  si  $x<0$}\\
{x} \text {  si  $x>=0$}
\end{cases}
\end{align}$

```python
x=np.linspace(-10,10)
y=np.piecewise(x, [x<0,x>=0],[lambda x:-x ,lambda x: x])
```

```python
plt.plot(x,y)
plt.show()
```
